```python
import pandas as pd
import difflib
import numpy as np
import sys

#file1 = "movie_list.txt"
file1 = sys.argv[1] 

#file2 = "movie_ratings.csv"
file2=sys.argv[2]

#file3 = "output.csv"
file3 = sys.argv[3] 


data = open(file1).readlines()
listofmovies = pd.DataFrame(data,columns=['title']) 
listofmovies['title'] = listofmovies['title'].replace('\n', '',regex=True)
#listofmovies
```

```python
movratings = pd.read_csv(file2, names=['title','rating'] )
#movratings
```

```python
def close_matches(word):
    value = difflib.get_close_matches(word, listofmovies['title'] ,n=500)
    return value
matches = np.vectorize(close_matches)
```

```python


movratings['title'] = movratings['title'].apply(matches)
movratings = movratings.explode('title')
#Dropping Null values and strings with no content
movratings = movratings.dropna()
movratings=movratings[movratings.title !='']
#Replacing . in strings with blank because of the csv file
movratings['title'] = movratings['title'].str.replace('.','')
movratings['rating'] =movratings['rating'].astype(int)
ratings=movratings.groupby('title').mean().round(2).reset_index()
ratings.to_csv(file3,index=False)
ratings
```
