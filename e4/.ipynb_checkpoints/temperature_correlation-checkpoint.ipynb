```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import sys
```

```python
file1 = sys.argv[1] 
#file1 = "stations.json.gz"
file2 = sys.argv[2] 
#file2 = "city_data.csv"
```

```python
stations = pd.read_json(file1, lines=True)
```

```python
stations['avg_tmax']=stations['avg_tmax']/10.0
```

```python
citiesdata=pd.read_csv(file2)
```

```python
citiesdata=citiesdata.dropna()
citiesdata
```

```python
citiesdata['area']=(citiesdata['area'])/(1000000)
```

```python
citiesdata = citiesdata[citiesdata['area']<10000]
```

```python
citiesdata
```

```python
#The citation below was found on:https://stackoverflow.com/questions/27928/calculate-distance-between-two-latitude-longitude-points-haversine-formula


def distance_between_points(latitude1,longitude1,latitude2,longitude2):
    R = 6371 # Radius of earth
    Lat = math.radians(latitude2-latitude1) 
    Lon = math.radians(longitude2-longitude1)
    ma=math.sin(Lat/2)
    mb=math.sin(Lat/2) 
    mc=math.cos(math.radians(latitude1))
    md=math.cos(math.radians(latitude2))
    me=math.sin(Lon/2)
    mf= math.sin(Lon/2)
    a = ma* mb + mc * md * me * mf
    c = 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))
    dis = R * c 
    return dis

funct = np.vectorize(distance_between_points)
```

```python
def distance(citiesdata, stations):
    a=citiesdata['latitude']
    b=citiesdata['longitude']
    c=stations['latitude']
    d=stations['longitude']
    stations['distance'] = funct(a,b,c,d )
    return stations['distance']

stations
```

```python
def best_tmax(citiesdata, stations):
    stations['distance'] = distance(citiesdata, stations)
    return stations['avg_tmax'][stations['distance'].idxmin()]
stations

```

```python
citiesdata['population density'] = citiesdata['population']/citiesdata['area']

```

```python
citiesdata['best_tmax'] = citiesdata.apply(best_tmax, 1, stations=stations)

```

```python
plt.plot(citiesdata['best_tmax'], citiesdata['population density'], 'b.')
plt.title('Relationship between Temperature and Population Density')
plt.xlabel('Avg Max Temperature (°C)')
plt.ylabel('Population Density (people/km²')
plt.show()
```
